A stationary Gaussian process $X(t)$ has power spectral density function:

\begin{equation*}
  S_X(j\omega)=\frac{2}{\omega^4 + 1}
\end{equation*}

1. Find $E(X)$ and $E(X^2)$

```python
from numpy import linspace
import matplotlib.pyplot as plt

omega = linspace(-5, 5, 100)
S = 2/(omega**4 + 1)

plt.plot(omega, S)
plt.title("Spectral density")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.grid()
plt.show()
```

```python
from sympy import symbols, integrate, pi, oo

S, omega = symbols('S omega')

S = 2/(omega**4 + 1)

EX = 1/(2*pi)*integrate(omega*S, (omega, -oo, oo))
display(EX)

EX2 = 1/(2*pi)*integrate(omega**2*S, (omega, -oo, oo))
display(EX2)
```
