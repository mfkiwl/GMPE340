A stationary Gaussian process $X(t)$ has power spectral density function:

\begin{equation*}
  S_X(j\omega)=\frac{2}{\omega^4 + 1}
\end{equation*}

1. Find $E(X)$ and $E(X^2)$

```python
from numpy import linspace
import matplotlib.pyplot as plt

omega = linspace(-5, 5, 100)
Sx = 2/(omega**4 + 1)

plt.plot(omega, Sx)
plt.title("Spectral density")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.grid()
plt.show()
```

```python
from sympy import symbols, integrate, pi, oo

omega = symbols('omega')

Sx = 2/(omega**4 + 1)

EX = 0 # No impulse function at the origin of Sx
display(EX)

EX2 = 1/(2*pi)*integrate(Sx, (omega, -oo, oo))
display(EX2)
```
