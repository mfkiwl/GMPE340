Noise measurements of a certain amplifier indicate that the rms output voltage is 100 uV. The frequency spectrum of the noise is constant from 0 to 10MHz and zero otherwise. (Hint: example 3.9)

1. Determine the spectral density function and the autocorrelation function

```python
from numpy import *
import matplotlib.pyplot as plt

x = linspace(-20, 20, 100)

# Probability Density Function
y1 = piecewise(x, [x<-10, (x>=-10) & (x<=10), x>10], [lambda x: 0, lambda x: 1/20 , lambda x: 0])
y2 = sin(x)/x
plt.plot(x, y1)
plt.title("Spectral Density Function")
plt.xlabel("MHz")
plt.ylabel("uV^2/MHz")
plt.grid()
plt.show()

plt.plot(x, y2)
plt.title("Autocorrelation Function")
plt.xlabel("1e-8s")
plt.ylabel("Correlation")
plt.grid()
plt.show()
```
