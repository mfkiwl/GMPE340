# Solution {-}

A common autocorrelation function encountered in physical problems is:

\begin{equation*}
  R(\tau)=\sigma^2 e^{-\beta |\tau|} \cos \omega_0 \tau
\end{equation*}

1. Find the corresponding spectral density function

\begin{equation*}
  S(j\omega)=\frac{\sigma^2 \beta}{(\omega+\omega_0)^2+\beta^2} + \frac{\sigma^2 \beta}{(\omega-\omega_0)^2+\beta^2}
\end{equation*}

2. Plot the autocorrelation and the spectral density

```python
from numpy import linspace, exp, cos
import matplotlib.pyplot as plt

sigma = 1
beta = 2
omega0 = 10

tau = linspace(-2, 2, 100)
R = sigma**2*exp(-beta*abs(tau))*cos(omega0*tau)

omega = linspace(-20, 20, 100)
S = sigma**2*beta/((omega + omega0)**2 + beta**2) + sigma**2*beta/((omega - omega0)**2 + beta**2)

# Plot autocorrelation
plt.plot(tau, R)
plt.title("Autocorrelation")
plt.xlabel("Lag")
plt.ylabel("Correlation")
plt.grid()
plt.show()

# Plot spectral density
plt.plot(tau, S)
plt.title("Spectral density")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.grid()
plt.show()
```
